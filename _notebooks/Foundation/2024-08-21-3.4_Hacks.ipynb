# Javascript Text Analyzer

```python
function textAnalyzer() {
    // Accept input from the user
    const originalString = prompt("Enter a sentence or multiple words:");
    
    // Display original string
    console.log("Original String: " + originalString);
    
    // Count total characters
    const totalCharacters = originalString.length;
    console.log("Total Characters (including spaces): " + totalCharacters);
    
    // Split the string into words
    const words = originalString.split(" ");
    
    // Find the longest word
    let longestWord = '';
    for (const word of words) {
        if (word.length > longestWord.length) {
            longestWord = word;
        }
    }
    console.log("Longest Word: " + longestWord + " (" + longestWord.length + " characters)");
    
    // Find the middle character
    const middleIndex = Math.floor(originalString.length / 2);
    const middleCharacter = originalString[middleIndex];
    console.log("Middle Character: " + middleCharacter);
    
    // Replace a word in the string
    const wordToReplace = prompt("Enter a word to replace:");
    const replacementWord = prompt("Enter the replacement word:");
    
    // Check if the word to replace exists and replace it
    let newString = originalString;
    if (newString.includes(wordToReplace)) {
        newString = newString.split(wordToReplace).join(replacementWord);
        console.log("Updated String: " + newString);
    } else {
        console.log(`The word "${wordToReplace}" is not in the original string.`);
    }
}

// Call the text analyzer function
textAnalyzer();
```

# Python Text Analyzer

```python
def text_analyzer():
    # Accept input from the user
    original_string = input("Enter a sentence or multiple words: ")
    
    # Display original string
    print("Original String:", original_string)
    
    # Count total characters
    total_characters = len(original_string)
    print("Total Characters (including spaces):", total_characters)
    
    # Split the string into words
    words = original_string.split()
    
    # Find the longest word
    longest_word = max(words, key=len) if words else ''
    print("Longest Word:", longest_word, "(", len(longest_word), "characters)")
    
    # Replace a word in the string
    word_to_replace = input("Enter a word to replace: ")
    replacement_word = input("Enter the replacement word: ")
    
    if word_to_replace in original_string:
        new_string = original_string.replace(word_to_replace, replacement_word)
        print("Updated String:", new_string)
    else:
        print(f'The word "{word_to_replace}" is not in the original string.')

# Call the text analyzer function
text_analyzer()
```
